```python
barishal_df_Yearly = df_loc_yearavg[df_loc_yearavg['Location'].str.startswith('bar')]
chittagong_df_Yearly = df_loc_yearavg[df_loc_yearavg['Location'].str.startswith('chi')]
cumilla_df_Yearly = df_loc_yearavg[df_loc_yearavg['Location'].str.startswith('cum')]
dhaka_df_Yearly = df_loc_yearavg[df_loc_yearavg['Location'].str.startswith('dha')]
gazipur_df_Yearly = df_loc_yearavg[df_loc_yearavg['Location'].str.startswith('gaz')]
khulna_df_Yearly = df_loc_yearavg[df_loc_yearavg['Location'].str.startswith('khu')]
mymensingh_df_Yearly = df_loc_yearavg[df_loc_yearavg['Location'].str.startswith('mym')]
narayanganj_df_Yearly = df_loc_yearavg[df_loc_yearavg['Location'].str.startswith('nar')]
narsingdi_df_Yearly = df_loc_yearavg[df_loc_yearavg['Location'].str.startswith('nars')]
rajshahi_df_Yearly = df_loc_yearavg[df_loc_yearavg['Location'].str.startswith('raj')]
rangpur_df_Yearly = df_loc_yearavg[df_loc_yearavg['Location'].str.startswith('ran')]
savar_df_Yearly = df_loc_yearavg[df_loc_yearavg['Location'].str.startswith('sav')]
sylhet_df_Yearly = df_loc_yearavg[df_loc_yearavg['Location'].str.startswith('syl')]


dhaka_df_Yearly
```

```python
barishal_df_Yearly['Date'] = pd.to_datetime(barishal_df_Yearly['Date'], errors = 'coerce', )
chittagong_df_Yearly['Date'] = pd.to_datetime(chittagong_df_Yearly['Date'], errors = 'coerce', )
cumilla_df_Yearly['Date'] = pd.to_datetime(cumilla_df_Yearly['Date'], errors = 'coerce', )
dhaka_df_Yearly['Date'] = pd.to_datetime(dhaka_df_Yearly['Date'], errors = 'coerce', )
gazipur_df_Yearly['Date'] = pd.to_datetime(gazipur_df_Yearly['Date'], errors = 'coerce', )
khulna_df_Yearly['Date'] = pd.to_datetime(khulna_df_Yearly['Date'], errors = 'coerce', )
mymensingh_df_Yearly['Date'] = pd.to_datetime(mymensingh_df_Yearly['Date'], errors = 'coerce', )
narayanganj_df_Yearly['Date'] = pd.to_datetime(narayanganj_df_Yearly['Date'], errors = 'coerce', )
narsingdi_df_Yearly['Date'] = pd.to_datetime(narsingdi_df_Yearly['Date'], errors = 'coerce', )
rajshahi_df_Yearly['Date'] = pd.to_datetime(rajshahi_df_Yearly['Date'], errors = 'coerce', )
rangpur_df_Yearly['Date'] = pd.to_datetime(rangpur_df_Yearly['Date'], errors = 'coerce', )
savar_df_Yearly['Date'] = pd.to_datetime(savar_df_Yearly['Date'], errors = 'coerce', )
sylhet_df_Yearly['Date'] = pd.to_datetime(sylhet_df_Yearly['Date'], errors = 'coerce', )




dhaka_df_Yearly.info()
```

```python
fig = plt.figure(figsize=(18,12))
sns.lineplot(data=barishal_df_Yearly, x='Date', y='AQImean', palette='deep', lw=5, alpha=0.7)
sns.scatterplot(data=barishal_df_Yearly, x='Date', y='AQImean', palette='deep', legend=False)
plt.xlim(datetime.strptime('2014-12-31', "%Y-%m-%d"), datetime.strptime('2021-12-31', "%Y-%m-%d"))

plt.title('Yearly Barishal Average AQI', fontsize = 25)
plt.ylabel('AQI', fontsize=30)
plt.xlabel('Date', fontsize=30)
plt.ylim(0,500)
```
